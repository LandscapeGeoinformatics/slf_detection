```python
import pandas as pd
import matplotlib.pyplot as plt
import glob
```

### Plot training and validation loss and accuracy

```python
df = pd.read_csv(
    "log.csv"
)
```

```python
df
```

```python
# with specific fontsize
plt.rcParams.update({'font.size': 16})

# create subplots (1 row, 2 columns)
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# loss
axes[0].plot(df["loss"], label="Train Loss")
axes[0].plot(df["val_loss"], label="Val Loss")
axes[0].set_xlabel("Epoch")
#axes[0].set_ylabel("BCE Loss")
axes[0].set_title("Binary Cross-Entropy loss")
axes[0].legend()
axes[0].grid(True)

# f1 score
axes[1].plot(df["f1"], label="Train F1-score")
axes[1].plot(df["val_f1"], label="Val F1-score")
axes[1].set_xlabel("Epoch")
#axes[1].set_ylabel("F1-score")
axes[1].set_title("F1-score")
axes[1].legend()
axes[1].grid(True)

plt.tight_layout(w_pad=3)
plt.show()
```

### Multiple csv plot

```python
# Load all CSV files matching the pattern
csv_files = glob.glob('log_*.csv')

# Initialize a dictionary to store dataframes
dataframes = {}

# Load each CSV file into a dataframe and store it in the dictionary
for file in csv_files:
    key = 'log_' + file.split('/')[-1].replace('.csv', '')
    df = pd.read_csv(file)
    # Cut the epoch
    df = df[df['epoch'] <= 5]
    dataframes[key] = df

# Plot val_loss for each dataframe
plt.figure(figsize=(8, 8))
for key, df in dataframes.items():
    plt.plot(df['epoch'], df['val_loss'], label=key)

plt.xlabel('Epoch')
plt.ylabel('Validation Loss')
plt.title('Validation Loss over Epochs')
plt.legend()
plt.show()
```

```python
# Load all CSV files matching the pattern
csv_files = glob.glob('log_*.csv')

# Initialize a dictionary to store dataframes
dataframes = {}

# Load each CSV file into a dataframe and store it in the dictionary
for file in csv_files:
    key = 'log_' + file.split('/')[-1].replace('.csv', '')
    df = pd.read_csv(file)
    # Cut the epoch to 30
    df = df[df['epoch'] <= 5]
    dataframes[key] = df

# Create a figure with two columns for the plots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 6))

# Plot loss for each dataframe in the first column
for key, df in dataframes.items():
    axes[0].plot(df['epoch'], df['loss'], label=key)
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Loss')
axes[0].set_title('Loss over Epochs')
axes[0].legend()

# Plot val_loss for each dataframe in the second column
for key, df in dataframes.items():
    axes[1].plot(df['epoch'], df['val_loss'], label=key)
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Validation Loss')
axes[1].set_title('Validation Loss over Epochs')
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
# Load all CSV files matching the pattern
csv_files = glob.glob('log_*.csv')

# Initialize a dictionary to store dataframes
dataframes = {}

# Load each CSV file into a dataframe and store it in the dictionary
for file in csv_files:
    key = 'log_' + file.split('/')[-1].replace('.csv', '')
    df = pd.read_csv(file)
    # Cut the epoch to 30
    df = df[df['epoch'] <= 5]
    dataframes[key] = df

# Create a figure with two columns for the plots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 6))

# Plot loss for each dataframe in the first column
for key, df in dataframes.items():
    axes[0].plot(df['epoch'], df['iou'], label=key)
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('IoU')
axes[0].set_title('IoU over Epochs')
axes[0].legend()

# Plot val_loss for each dataframe in the second column
for key, df in dataframes.items():
    axes[1].plot(df['epoch'], df['val_iou'], label=key)
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Validation IoU')
axes[1].set_title('Validation IoU over Epochs')
axes[1].legend()

plt.tight_layout()
plt.show()
```
